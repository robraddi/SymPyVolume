<h1> SymPyVolume</h1>
<p>Command line 3-D Plotting tool derived from <a href="https://www.sympy.org/en/index.html" target="_blank">SymPy</a> and <a href="https://ipyvolume.readthedocs.io/en/latest/" target="_blank">IPyVolume</a>.</p>

<h2> Still a work in progress...</h2>

<h1> Installation</h1>

<ul>
    <li style="list-style-type: none;font-weight: normal;"><b>1.</b> Clone the Repo to your github directory </li>
    <li style="list-style-type: none;font-weight: normal;"><b>2.</b> Create an alias in bashrc for the plot.py script:</li>
</ul>


<!-- language: bash -->
<code>alias plot="python "$home"/github/SymPyVolume/plot.py'</code>


<p>You're all set. Checkout the help page for more information</p>


<h3>Usage:</h3>



<!-- language: bash -->
<code style="language: bash;">
optional arguments:
  -h, --help            show this help message and exit
  --Fn <class 'str'>    Function with variables x,y,z
  --x X                 x-axis limits
  --y Y                 y-axis limits
  --z Z                 z-axis limits
  --axis-labels AXIS_LABELS
                        axis labels for plot
  --data DATA           input data file
</code>


<hr>

<h1>Examples:</h1>


<p>Let's start with something basic:</p>

<p>Here's a very simple plot of $x/y$, where $0 < x < 10$, and where $0 < y < 10$</p>

<!-- language: bash -->
<code>plot --Fn "x/y" --x "[0,10]" --y "[0,10]"</code>

<iframe src="images/xFracy.html" width="500" height="500" align="center"></iframe>

<!-- language: bash -->
<code>plot --Fn "sin(x)*cos(y)" --x "[0,10" --y "[0,10]"</code>

<iframe src="images/sin{x}dotcos{y}.html" width="500" height="500" align="center"></iframe>


<p>We can even add more than one plot in the same figure by seperating the functions by a semicolon.</p>


<!-- language: bash -->
<code>plot --Fn "sin(y)*sin(x);0.01*x-0.1*y" --x "[0,10]" --y "[0,10]"</code>


<iframe src="images/sin{y}dotsin{x}_and_0.01dotx-0.1doty.html" width="500" height="500" align="center"></iframe>


<p>You can even add axis labels...</p>


<!-- language: bash -->
<code>plot --Fn "4*pi*x**(2)*(.001/(2*pi*y))**(3/2)*exp(-.001*x**(2)/(2*y))" --axis-labels "\nu(m/s),T(K),F(\nu)(s/m)" --x "[0,1000]" --y "[0,1000]" --z "[0,0.0095]"</code>
